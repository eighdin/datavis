```python
import nltk
import random
nltk.download('punkt_tab')
```

```python
sentence = "I like watching Nolan's, Spielbergs, and Cameron's movies"
tokens = nltk.word_tokenize(sentence)
bigrams = list(nltk.bigrams(tokens))
index_to_flip = random.randint(0, len(bigrams)-1)
flipped_bigram = [bigrams[index_to_flip][1], bigrams[index_to_flip][0]]
print(flipped_bigram)
```

```python
flipped_sentence = tokens[:index_to_flip] + flipped_bigram + tokens[index_to_flip+2:]
flipped_sentence = ' '.join(flipped_sentence)
print(flipped_sentence)
```
